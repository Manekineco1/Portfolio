### Напишите функцию для вычисления итоговой sMAPE

<b/> Удалим отсутствующие в тестовой выборке признаки (за исключением целевых признаков) из тренировочной выборки. В противном случае в ходе обучения будущих моделей может произойти утечка целевого признака. Она происходит, когда в признаки попадает информация о целевом признаке. <b>

```python
features_to_delete = missed_features - set(['rougher.output.recovery','final.output.recovery'])
```

```python
print('Количество признаков для удаления из тренировочной выборки -', len(features_to_delete))
```

```python
df_train = df_train.drop(columns=features_to_delete, axis=1) #Удалим колонки с этими 33 признаками
```

```python
#Напишем функцию для вычисления sMape
def get_smape(target, predict):
    smape = (1/len(target))*np.sum((np.abs(target - predict))/( (np.abs(target)+np.abs(predict))/2))*100
    return smape
```

```python
smape_scorer = make_scorer(get_smape, greater_is_better = False)
```

```python
#напишем функцию для вычисления итоговой sMAPE
def get_final_smape(rougher, final):
    final = 0.25*rougher+0.75*final
    return final
```

 - Из тестовой выборки были удалены отсуствующие в тренировочной выборке признаки, во избежание утечки данных.
 - Была написала функция для вычисления показателей sMAPE и итоговое sMAPE.
